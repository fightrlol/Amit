```python
class Factorial:
    """
    Factorial class calculates the factorial of a given number.
    
    Attributes:
        num (int): The number to calculate the factorial for.
    """
    
    def __init__(self, num):
        self.num = num

    def calculate(self):
        """
        Recursively calculates the factorial of the number.
        
        Returns:
            int: Factorial of the number.
        """
        if self.num == 0 or self.num == 1:
            return 1
        return self.num * Factorial(self.num - 1).calculate()



fact_obj = Factorial(5)
print(fact_obj.calculate())
```

```python
class PrimeCheck:
    """
    PrimeCheck class checks whether a number is prime.
    
    Attributes:
        num (int): The number to be checked.
    """
    
    def __init__(self, num):
        """
        Initializes the PrimeCheck object with a number.
        
        Args:
            num (int): The number to check for primality.
        """
        self.num = num

    def check(self):
        """
        Determines whether the number is prime.
        
        Returns:
            bool: True if the number is prime, False otherwise.
        """
        if self.num < 2:
            return False
        for i in range(2, self.num):
            if self.num % i == 0:
                return False
        return True


prime_obj = PrimeCheck(20)
print(prime_obj.check())
```

```python
class NumberOperations:
    """
    NumberOperations class handles finding common divisors of two numbers
    and generating a list of even numbers up to a given number.
    
    Attributes:
        num1 (int): The first number for common divisors.
        num2 (int): The second number for common divisors.
        limit (int): The limit up to which even numbers are generated.
    """
    
    def __init__(self, num1, num2, limit):
        """
        Initializes the NumberOperations object with two numbers and a limit.
        
        Args:
            num1 (int): The first number for divisors.
            num2 (int): The second number for divisors.
            limit (int): The upper limit for even numbers.
        """
        self.num1 = num1
        self.num2 = num2
        self.limit = limit

    def common_divisors(self):
        """
        Finds all common divisors of num1 and num2.
        
        Returns:
            list: A list of common divisors.
        """
        mn = min(self.num1, self.num2)
        common_divs = []
        for i in range(1, mn + 1):
            if self.num1 % i == 0 and self.num2 % i == 0:
                common_divs.append(i)
        return common_divs

    def even_numbers(self):
        """
        Generates a list of even numbers up to the specified limit.
        
        Returns:
            list: A list of even numbers.
        """
        evens = []
        for i in range(self.limit):
            if i % 2 == 0:
                evens.append(i)
        return evens



num_ops = NumberOperations(20, 10, 10)
print(num_ops.common_divisors())
print(num_ops.even_numbers())
```

```python
class LongestCommonSubstring:
    """
    LongestCommonSubstring class finds the longest common substring between two strings.
    
    Attributes:
        str1 (str): The first string.
        str2 (str): The second string.
    """
    
    def __init__(self, str1, str2):
        """
        Initializes the LongestCommonSubstring object with two strings.
        
        Args:
            str1 (str): The first string.
            str2 (str): The second string.
        """
        self.str1 = str1
        self.str2 = str2

    def find(self):
        """
        Finds the longest common substring between str1 and str2.
        
        Returns:
            str: The longest common substring.
        """
        longest = ""
        for i in range(len(self.str1)):
            for j in range(i, len(self.str1)):
                sub_str = self.str1[i:j+1]
                if sub_str in self.str2 and len(sub_str) > len(longest):
                    longest = sub_str
        return longest


lcs_obj = LongestCommonSubstring(input(), input())
print(lcs_obj.find())
```

```python
class ReverseWords:
    """
    ReverseWords class reverses the order of words in a given string.
    
    Attributes:
        s (str): The string to reverse.
    """
    
    def __init__(self, s):
        """
        Initializes the ReverseWords object with a string.
        
        Args:
            s (str): The string whose words are to be reversed.
        """
        self.s = s

    def reverse(self):
        """
        Reverses the order of words in the string.
        
        Returns:
            str: The string with words in reversed order.
        """
        lst = self.s.split(" ")
        new_lst = lst[::-1]
        return " ".join(new_lst)


rev_obj = ReverseWords("Hello World")
print(rev_obj.reverse())
```

```python
import random

class PasswordGen:
    """
    PasswordGen class generates a random password of a specified length.
    
    Attributes:
        length (int): The length of the password to generate.
    """
    
    def __init__(self, length):
        """
        Initializes the PasswordGen object with a password length.
        
        Args:
            length (int): Length of the password.
        """
        self.length = length

    def generate_password(self):
        """
        Generates a random password using letters, digits, and symbols.
        
        Returns:
            str: The generated random password.
        """
        characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
        num_of_chars = len(characters)
        password = ""
        for _ in range(self.length):
            rand_indx = random.randrange(0, num_of_chars)
            password += characters[rand_indx]
        return password


class Person:
    """
    Person class represents a person with a color and a movement length.
    
    Attributes:
        color (str): The color of the person.
        length (int): The movement length.
    """
    
    def __init__(self, color, length):
        """
        Initializes the Person object with color and length.
        
        Args:
            color (str): The color of the person.
            length (int): The movement length.
        """
        self.color = color
        self.length = length

    def run(self):
        """
        Prints a message indicating movement and the person's color.
        """
        print(f"I am moving and my color is {self.color}")



if __name__ == "__main__":
    p1 = PasswordGen(8)
    print(p1.generate_password())
    
    p = Person('red', 10)
    p.run()
```

```python
class WordCounter:
    """
    WordCounter class counts the frequency of each unique word in a list.
    
    Attributes:
        words (list): The list of words to count.
    """
    
    def __init__(self, words):
        """
        Initializes the WordCounter object with a list of words.
        
        Args:
            words (list): The list of words.
        """
        self.words = words

    def count_words(self):
        """
        Counts the frequency of each unique word in the list.
        
        Returns:
            dict: A dictionary with words as keys and their counts as values.
        """
        unique_words = set(self.words)
        counts = {word: 0 for word in unique_words}
        for i in unique_words:
            for j in self.words:
                if i == j:
                    counts[i] += 1
        return counts



words_list = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]
wc = WordCounter(words_list)
print(wc.count_words())
```

```python
import random

class ChatBot:
    """
    ChatBot class simulates a simple chatbot with predefined responses.
    
    Attributes:
        responses (dict): A dictionary mapping keywords to possible responses.
    """
    
    def __init__(self):
        """
        Initializes the ChatBot object with predefined responses.
        """
        self.responses = {
            "hello": ["Hello!", "Hi there!", "Greetings!"],
            "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
            "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
            "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
        }

    def get_response(self, user_input):
        """
        Generates a response based on user input.
        
        Args:
            user_input (str): The input text from the user.
            
        Returns:
            str: A response from the chatbot.
        """
        for key in self.responses:
            if key in user_input:
                return random.choice(self.responses[key])
        return random.choice(self.responses["default"])

    def start(self):
        """
        Starts the chatbot conversation.
        """
        print("Chatbot: Hi! How can I assist you today?")
        while True:
            user_input = input("User: ").lower()
            response = self.get_response(user_input)
            print("Chatbot:", response)
            if user_input == "goodbye":
                break


bot = ChatBot()
bot.start()
```
