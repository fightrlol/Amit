```python
def factorial(num):
    if num == 0:
        return 0
    if num == 1:
        return 1
    return num * factorial(num -1)

factorial(1)
```

```python
def isPrime(num):
    for i in range(2,num):
        if num%i == 0:
            return False
    return True
    
isPrime(20)    
```

```python
def dividors(num1,num2):
    mn = min(num1,num2)
    common_dividors = []
    for i in range(1,mn+1):
        if num1%i == 0 and num2%i ==0:
            common_dividors.append(i)
    return  common_dividors  
```
